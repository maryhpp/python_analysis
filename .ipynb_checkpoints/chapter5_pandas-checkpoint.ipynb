### 索引对象

pandas的索引对象负责管理轴标签和其他元数据（⽐如轴名称 等）。构建Series或DataFrame时，所⽤到的任何数组或其他序 列的标签都会被转换成⼀个Index

```python
obj = pd.Series(range(3), index=['a', 'b', 'c'])
index = obj.index
index
```

```python
index[1:]
```

```python
index[1] = 'd'  #index对象是不可变的，因此用户不能对其进行修改
```

不可变可以使Index对象在多个数据结构之间安全共享： 

```python
labels = pd.Index(np.arange(3))
labels
```

```python
obj2 = pd.Series([1.5, -2.5, 0], index=labels)
obj2
```

```python
obj2.index is labels
```

```python
frame3
```

```python
frame3.columns
```

```python
'Ohio' in frame3.columns
```

```python
2003 in frame3.index
```

## 基本功能

### 重新索引

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
```

```python
#用该Series的reindex将会根据新索引进行重排。如果某个索引值当前不存在，就引入缺失值
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])  
obj2
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
```

```python
obj3.reindex(range(6), method='ffill')  #ffill可以实现前向值填充
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California'])
frame
```

```python
frame2 = frame.reindex(['a', 'b', 'c', 'd'])  #reindex可以修改行索引和列。只传递一个序列是，会重新索引结果的行
frame2
```

```python
#列可以使用columns关键字重新索引
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states)
```

默认情况下，DataFrame和Series之间的算术运算会将Series的 索引匹配到DataFrame的列，然后沿着⾏⼀直向下⼴播： 

```python
frame - series
```

如果某个索引值在DataFrame的列或Series的索引中找不到，则 参与运算的两个对象就会被重新索引以形成并集：


```python
frame + series2
```

如果你希望匹配⾏且在列上⼴播，则必须使⽤算术运算⽅法。

```python
series3 = frame['d']
frame
```

```python
frame.sub(series3, axis='index')
```

### 函数应用和映射

```python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
np.abs(frame)
```

另⼀个常⻅的操作是，将函数应⽤到由各列或⾏所形成的⼀维数 组上。DataFrame的apply⽅法即可实现此功能

```python
#这里的函数f，计算了一个Series的最大和最小的差，在frame的每列都执行了一次，结果是一个Series，使用frame的列作为索引
f = lambda x: x.max() - x.min()  
frame.apply(f)    
```

```python
frame
```

```python
#如果axis='columns'到apply,这个函数会在每行执行
frame.apply(f, axis='columns')
```

传递到apply的函数不是必须返回⼀个标量，还可以返回由多个 值组成的Series

```python
def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
frame.apply(f)
```

元素级的Python函数也是可以⽤的。假如你想得到frame中各个 浮点值的格式化字符串，使⽤applymap即可

```python
format = lambda x: '%.2f' %x
frame.applymap(format)
```

```python
#之所以叫做applymap，是因为Series有一个应用到元素级函数的map方法
frame['e'].map(format)
```

### 唯一值、值计算以及成员资格

```python
obj = pd.Series(['c', 'a', 'd', 'a', 'a', 'b', 'b', 'c', 'c'])
```

```python
uniques = obj.unique()  #得到Series中的唯一值数组，是未排序的
uniques
```

```python
obj.value_counts()
```

```python
#value_counts还是一个顶级pandas方法，可用于任何数组或序列
pd.value_counts(obj.values, sort=False)
```

isin⽤于判断⽮量化集合的成员资格，可⽤于过滤Series中或 DataFrame列中数据的⼦集： 

```python
obj
```

```python
mask = obj.isin(['b', 'c'])
mask
```

```python
obj[mask]
```

与isin类似的是Index.get_indexer⽅法，它可以给你⼀个索引数 组，从可能包含重复值的数组到另⼀个不同值的数组

```python
to_match = pd.Series(['c', 'a', 'b', 'b', 'c', 'a'])
unique_vals = pd.Series(['c', 'b', 'a'])
pd.Index(unique_vals).get_indexer(to_match)
```

```python
data = pd.DataFrame({'Qu1': [1, 3, 4, 3, 4],
                     'Qu2': [2, 3, 1, 2, 3],
                     'Qu3': [1, 5, 2, 4, 4]})
data
```

```python
result = data.apply(pd.value_counts).fillna(0)
result
```

```python
data.apply(pd.value_counts)
```
